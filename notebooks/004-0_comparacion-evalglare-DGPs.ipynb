```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dgp_def = pd.read_csv('../resultados/dgp_default.txt', sep = " ", header = None)
```

```python
cols = dgp_def[0][0].split(sep = ",")
dgp_def.drop([0,7,8], axis = 1, inplace = True)
dgp_def.columns = cols
dgp_def
```

```python
dgp_d2000 = pd.read_csv('../resultados/dgp_d2000.txt', sep = " ", header = None)
dgp_d2000.drop([0,7,8], axis = 1, inplace = True)
dgp_d2000.columns = cols
dgp_d2000
```

```python
dgp_t = pd.read_csv('../resultados/dgp_t.txt', sep = " ", header = None)
dgp_t.drop([0,7,8], axis = 1, inplace = True)
dgp_t.columns = cols
dgp_t
```

```python
dgps = pd.read_csv('../resultados/DGPs/DGPu_A10-1_anual.ill', sep = " ", skiprows = range(0,10), header = None, skipinitialspace = True)
```

```python
DGPs = dgps.T * 6.22 * 10**-5 + 0.184
DGPs.index = pd.date_range('2019-01-01 00:30:00', periods=8760, freq='H')
```

```python
DGPs['2019-09-29 12'].T.plot(label = "DGPs", marker = "*")
dgp_t.dgp.iloc[1:].plot(label = "t", marker = "*")
dgp_d2000.dgp.iloc[1:].plot(label = "b = 2000 cd/m2", marker = "*")
dgp_def.dgp.iloc[1:].plot(label = "default", linestyle = "dotted", marker = "*")
plt.legend()
plt.ylim(0.2,0.59)
plt.xlim(0,12)
plt.hlines(0.352,0, 12, linewidth= 0.5)
plt.hlines(0.394,0, 12, linewidth= 0.5)
plt.hlines(0.456,0, 12, linewidth= 0.5)
```
