```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# importar datos generados con evalglare, mismo script
DGP = pd.read_csv('../resultados/revision_Ib/glare.txt', sep = " ",header = None)
```

```python
# Formato a dataframe
DGP.dropna(axis=1,inplace=True)
DGP['Aula'] = (['A10-1']*4)+(['A10-2']*4)+(['A20-1']*3)+(['A20-2']*3)+(['A40-1']*7)+(['A40-2']*7)
sur0 = ['22/12/2019 00:09:00','12/12/2019 00:09:00','22/11/2019 00:09:00','12/11/2019 00:09:00']
sur = sur0[:]
norte = ['20/06/2019 00:18:00','15/06/2019 00:18:00','22/05/2019 00:18:00']
sur.extend(sur)
sur.extend(norte*2)
sur.extend(sur0)
sur.extend(norte)
sur.extend(sur0)
sur.extend(norte)
DGP['Fecha'] = sur
DGP['Fecha'] = pd.to_datetime(DGP['Fecha'])
DGP = DGP.pivot(index='Aula',columns='Fecha')[1]
```

```python
DGP
```

```python
# Graficas fachada sur
Ax2 = DGP.T['2019-11-22': '2019-12-22'].T.plot.bar(cmap = 'Accent',width=0.85, figsize = (9,6), grid = True, rot = 0)
Ax2.set_ylim(0,0.5)
Ax2.set_ylabel('DGP')
Ax2.hlines([0.35,0.39,0.45],-1,6, colors = ['green','goldenrod','orangered'], linewidth = 1)
Ax2.set_title('DGP para aulas con ventanas en fachada sur')
plt.savefig('../resultados/revision_Ib/glare/sur.png')
```

```python
# Graficas fachada norte
Ax1 = DGP.T['2019-05-22': '2019-06-20'].T.plot.bar(cmap = 'tab10', width=0.85, figsize = (9,6), grid = True, rot = 0)
Ax1.set_ylim(0,0.5)
Ax1.hlines([0.35,0.39,0.45],-1,6, colors = ['green','goldenrod','orangered'], linewidth = 1)
Ax1.set_title('DGP para aulas con ventanas en fachada norte')
Ax1.set_ylabel('DGP')
plt.savefig('../resultados/revision_Ib/glare/norte.png')
```
